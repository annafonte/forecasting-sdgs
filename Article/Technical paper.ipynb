### Introduction

<p style='text-align: justify;'> The 2030 Agenda and the Sustainable Development Goals (SDGs) were developed at the United Nations Conference on Sustainable Development, held in Rio de Janeiro in 2012, and replace the old Millennium Development Goals (MDGs), with which a global initiative for tackling poverty was undertaken.
The 2030 Agenda is a comprehensive, universal agenda that refers to the three dimensions of sustainable development (economic, social and environmental). This global framework is based on five pillars: people, the planet, prosperity, peace and alliances, and is deployed through a system of 17 SDGs and 115 indicators, through which it is proposed to address the major global challenges, from the fight against poverty or climate change to education, health, gender equality, peace or sustainable cities. </p>

### Literature review

<h4 style='text-align: justify;'> The SDGs Goal Challenge </h4>

<p style='text-align: justify;'> Many studies on forecasting SDGs and related challenges have been published. To monitor the progress of SDGs, the UN publishes a yearly report (UN, 2019) to measure the progress towards the global attainment of the SDGs and provides a good annual general overview. The UN also publishes statistics used to monitor progress towards SDG attainment (this is the input data used with respect to the proposed framework). 
The majority of the available literature has focused on forecasting individual SDGs as opposed to the whole set of SDGs as proposed by the UN.</p>

<h4 style='text-align: justify;'> Time Series Forecasting </h4>

<p style='text-align: justify;'> Time series analysis has been the subject of much research. Much of this
work has been directed at supervised learning, the mapping of time series to class labels of some kind. In the context
of SDGs prediction, a particular challenge is the nature of the time series data available; at time of writing (2020) this was limited to 5 observation points per time series.
Any forecasting method, considered in the context of the proposed framework, must therefore be able to operate using such short time series. From the literature there are three models that seem appropriate: Auto-Regressive Moving Average-ARMA (Lawrance and Lewis, 1980), AutoRegressive Integrated Moving Average-ARIMA (Hyndman,
2018), and Facebook Prophet-Fbprophet (Sean J and Benjamin, 2017). </p>

### The SDG dataset

<p style='text-align: justify;'> As noted above, the UN identified 17 SDGs. Each SDG has between 3 and 13 targets, and each target, in turn, has a number of indicators associated with it. The SDG data set is publicly available from the SDG website. At time of writing (2020) the data set spanned an 5 year period (2015-2020). The SDG dataset is relatively large, and is comprised of arround 100,000 records holding statistical SDG information covering 195 countries. </p>

<p style='text-align: justify;'>  The global SDG Index score can be interpreted as the percentage of achievement. The difference between 100 and countries’ scores is therefore the distance in percentage that needs to be completed to achieving the SDGs and goals. The same basket of indicators is used for all countries to generate comparable scores and rankings. It should be
noted that differences in rankings may be due to small differences in the aggregate score. The SDG Index score signifies a country's position between the worst (0) and the best or target (100) outcomes. </p>

<p style='text-align: justify;'> At the indicator level, various procedures were in place to address missing data biases:: </p>
    
- <p style='text-align: justify;'> First, the data was scaled and transformed in order to make the it comparable across indicators. Each variable was rescaled from 0 to 100 with 0 denoting worst performance and 100 describing the technical optimum. Rescaling is usually very sensitive to the choice of limits and extreme values (outliers) at both tails of the distribution. Each indicator distribution was censored, so that all values exceeding the upper bound scored 100, and values below the lower bound scored 0. ### Results

<p style='text-align: justify;'> Different analysis were performed in order to understand the performance of the SDGs. First, a timeseries forecasting model was applied in order to find out if the Agenda 2030 can be achieved before the deadline. Then, an analysis on each SDGs performance was conducted. Thirdly, the relationship between the income of the different countries and the achievement of the SDGs was explored. Finally, an analysis regarding the geographical regions of the UN was also performed. </p>

#### The SDG Index prediction framework

<p style='text-align: justify;'> In order to find out if the SDG Index will be globally meet in 2030, a timeseries forecasting analysis was performed, as noted above. The red line is forecasting SDG Index achievement and the area between the orange and blue line is the 95% confidence interval around the forecast.The figure below shows that the SDG Index will not be met in the world, which explains that we need to step up our collective efforts to reach the goals. </p>

<p style='text-align: justify;'> Moreover, the model was validated by training the model with different time series data. For the first validation, where the model was trained with data from 5 years, the R2 Score had a value of 0.86, and  less accuracy when the model was trained with less data. To conclude, it can be confirmed that in order to work with historical data, the more data collected, the more accuracy will have the model. However, the recent context of the study topic explains the lack of historical data regarding the SDGs.</p>

#### Global SDG performance by goal

<p style='text-align: justify;'> For certain goals, good performance is proved, especially for SDGs 12 & 13: Responsable consumption and production and Climate action, which appear to be in above the 75th percentile. However, SDGs 17 and 9, which are above 50th percentile of performance, emphasize the need to foster innovations and policies that can make our industries and infrastructures more sustainable, as well as the need to improve the partnership between countries to achieve the goals. The alarming inability to achieve some of the goals, and indeed the deterioration for some SDGs, can seriously threaten the achievement of the Agenda 2030. </p>

#### Progress on the SDG Index by income group (2015–2020)

<p style='text-align: justify;'>  Looking at trends, many high income countries are not making significant progress
on SDG achievement, particularly in relation to other income groups. On average, progress
since 2015 has been faster in low- and middle-income countries compared with high-income countries. Covid-19 will likely negatively impact progress towards most SDGs in the short and medium-term, including in high-income countries.
Low-income countries tend to have lower SDG Index scores. This may be due to the nature of the SDGs, which focus to a large extent on ending extreme poverty and on access to basic services and infrastructure (SDGs 1–9). Moreover, poorer countries tend to lack adequate infrastructure and mechanisms to manage key environmental challenges covered under SDGs 12–15. </p>
